```python
import sys
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
sys.path.insert(1, os.path.join(sys.path[0], '/Users/henryxu/Desktop/Research/prismio'))
from prismio.io_frame import IOFrame
from prismio.io_frame_plotter import IOFramePlotter
import seaborn as sns
```

### load haccio runtime dataframe from csv

```python
original_haccio = pd.read_csv('/Users/henryxu/Desktop/Research/profiles/app-stdio-csv/haccio.csv')
original_haccio = original_haccio.groupby(['scaling', 'num_proc', 'io_subsystem', 'iteration']).agg({'time':'sum'})
original_haccio.reset_index(inplace=True)

original_haccio_avg = original_haccio.groupby(['scaling', 'num_proc', 'io_subsystem']).agg({'time':'mean'})
original_haccio_avg.reset_index(inplace=True)
```

### plot haccio runtime over all io subsystems

```python
haccio = original_haccio.copy()
haccio_avg = original_haccio_avg.copy()
haccio = haccio[~haccio['io_subsystem'].str.contains('cori')].copy()
haccio_avg = haccio_avg[~haccio_avg['io_subsystem'].str.contains('cori')].copy()

scalings = ['strong-scaling', 'weak-scaling']
for scaling in scalings:
    haccio_avg_scaling = haccio_avg[haccio_avg['scaling'] == scaling].copy()
    plt.figure()
    sns.lineplot(x='num_proc', y='time', hue='io_subsystem', data=haccio_avg_scaling, style="io_subsystem", markers=True, dashes=False)
    plt.title('haccio ' + scaling + ' runtime over all io subsystems')
    plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
    plt.savefig('/Users/henryxu/Desktop/io-project/figs/pre-analysis' + '/haccio-' + scaling + '-runtime-all-io-subsystems.pdf', bbox_inches='tight')

    haccio_scaling = haccio[haccio['scaling'] == scaling].copy()
    plt.figure()
    sns.lineplot(x='num_proc', y='time', hue='io_subsystem', data=haccio_scaling, ci=100, style="io_subsystem", markers=True, dashes=False)
    plt.title('haccio ' + scaling + ' runtime over all io subsystems')
    plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
    plt.savefig('/Users/henryxu/Desktop/io-project/figs/pre-analysis' + '/haccio-' + scaling + '-runtime-all-io-subsystems-tsplot.pdf', bbox_inches='tight')
    
    plt.figure()
    sns.pointplot(x='num_proc', y='time', hue='io_subsystem', data=haccio_scaling, ci=100, capsize=.15, errwidth=2.5, style="io_subsystem")
    plt.title('haccio ' + scaling + ' runtime over all io subsystems')
    plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
    plt.savefig('/Users/henryxu/Desktop/io-project/figs/pre-analysis' + '/haccio-' + scaling + '-runtime-all-io-subsystems-errbar.pdf', bbox_inches='tight')
```

### load ior runtime dataframe from csv

```python
original_ior = pd.read_csv('/Users/henryxu/Desktop/Research/profiles/app-stdio-csv/ior.csv')

original_ior = original_ior.groupby(['scaling', 'io_subsystem', 'num_proc', 'transfer_size', 'iteration']).agg({'time':'sum'})
original_ior.reset_index(inplace=True)

original_iort4k = original_ior[original_ior['transfer_size'] == '4k'].copy()
original_iort64k = original_ior[original_ior['transfer_size'] == '64k'].copy()
original_iort1m = original_ior[original_ior['transfer_size'] == '1m'].copy()

original_ior_avg = original_ior.groupby(['scaling', 'io_subsystem', 'num_proc', 'transfer_size']).agg({'time':'mean'})
original_ior_avg.reset_index(inplace=True)

original_iort4k_avg = original_ior_avg[original_ior_avg['transfer_size'] == '4k'].copy()
original_iort64k_avg = original_ior_avg[original_ior_avg['transfer_size'] == '64k'].copy()
original_iort1m_avg = original_ior_avg[original_ior_avg['transfer_size'] == '1m'].copy()
```

### plot ior runtime over all default io subsystems

```python
ior = original_ior.copy()
ior_avg = original_ior_avg.copy()
ior = ior[~ior['io_subsystem'].str.contains('cori')].copy()
ior_avg = ior_avg[~ior_avg['io_subsystem'].str.contains('cori')].copy()

transfer_sizes = ['4k', '64k', '1m']
scalings = ['strong-scaling', 'weak-scaling']
for transfer_size in transfer_sizes:
    ior_transfer_size = ior[ior['transfer_size'] == transfer_size].copy()
    ior_avg_transfer_size = ior[ior['transfer_size'] == transfer_size].copy()
    
    for scaling in scalings:
        ior_avg_transfer_size_scaling = ior_avg_transfer_size[ior_avg_transfer_size['scaling'] == scaling].copy()
        plt.figure()
        sns.lineplot(x='num_proc', y='time', hue='io_subsystem', ci=100, data=ior_avg_transfer_size_scaling, style="io_subsystem", markers=True, dashes=True)
        plt.title('iort' + transfer_size + ' ' + scaling + ' runtime over all default io subsystems')
        plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
        plt.savefig('/Users/henryxu/Desktop/io-project/figs/pre-analysis' + '/ior-t' + transfer_size + '-' + scaling + '-runtime-all-default-io-subsystems.pdf', bbox_inches='tight')
    
        ior_transfer_size_scaling = ior_transfer_size[ior_transfer_size['scaling'] == scaling].copy()
        plt.figure()
        sns.lineplot(x='num_proc', y='time', hue='io_subsystem', ci=100, data=ior_transfer_size_scaling, style="io_subsystem", markers=True, dashes=True)
        plt.title('iort' + transfer_size + ' ' + scaling + ' runtime over all default io subsystems')
        plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
        plt.savefig('/Users/henryxu/Desktop/io-project/figs/pre-analysis' + '/ior-t' + transfer_size + '-' + scaling + '-runtime-all-default-io-subsystems-tsplot.pdf', bbox_inches='tight')
    
        plt.figure()
        sns.pointplot(x='num_proc', y='time', hue='io_subsystem', ci=100, data=ior_transfer_size_scaling, style="io_subsystem", capsize=.15, errwidth=2.5)
#         ax.get_legend().remove()
        plt.title('iort' + transfer_size + ' ' + scaling + ' runtime over all default io subsystems')
        plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
        plt.savefig('/Users/henryxu/Desktop/io-project/figs/pre-analysis' + '/ior-t' + transfer_size + '-' + scaling + '-runtime-all-default-io-subsystems-errbar.pdf', bbox_inches='tight')
```
