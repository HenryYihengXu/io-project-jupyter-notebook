```python
import sys
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
sys.path.insert(1, os.path.join(sys.path[0], '/Users/henryxu/Desktop/Research/prismio'))
from prismio.io_frame import IOFrame
from prismio.io_frame_plotter import IOFramePlotter
import seaborn as sns
```

### load haccio runtime dataframe from csv

```python
haccio = pd.read_csv('/Users/henryxu/Desktop/Research/profiles/app-stdio-csv/haccio.csv')
```

```python
haccio = haccio.groupby(['scaling', 'num_proc', 'io_subsystem', 'iteration']).agg({'time':'sum'})
haccio.reset_index(inplace=True)
haccio = haccio.groupby(['scaling', 'num_proc', 'io_subsystem']).agg({'time':'mean'})
haccio.reset_index(inplace=True)
haccio
```

### plot haccio runtime over all io subsystems

```python
scalings = ['strong-scaling', 'weak-scaling']
for scaling in scalings:
    haccio_scaling = haccio[haccio['scaling'] == scaling].copy()
    plt.figure()
    sns.lineplot(x='num_proc', y='time', hue='io_subsystem', data=haccio_scaling, style="io_subsystem", markers=True, dashes=False)
    plt.title('haccio ' + scaling + ' runtime over all io subsystems')
    plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
    plt.savefig('/Users/henryxu/Desktop/paper/papers-2021-prismio/jupyter-notebook/plots/haccio/' + 'haccio-' + scaling + '-runtime-all-io-subsystems.pdf', bbox_inches='tight')
```

### load ior runtime dataframe from csv

```python
ior = pd.read_csv('/Users/henryxu/Desktop/Research/profiles/app-stdio-csv/ior.csv')
```

```python
ior = ior.groupby(['scaling', 'io_subsystem', 'num_proc', 'transfer_size', 'iteration']).agg({'time':'sum'})
ior.reset_index(inplace=True)
ior = ior.groupby(['scaling', 'io_subsystem', 'num_proc', 'transfer_size']).agg({'time':'mean'})
ior.reset_index(inplace=True)
iort4k = ior[ior['transfer_size'] == '4k'].copy()
iort64k = ior[ior['transfer_size'] == '64k'].copy()
iort1m = ior[ior['transfer_size'] == '1m'].copy()
ior
```

### plot ior runtime over all default io subsystems

```python
transfer_sizes = ['4k', '64k', '1m']
scalings = ['strong-scaling', 'weak-scaling']
for transfer_size in transfer_sizes:
    ior_transfer_size = ior[ior['transfer_size'] == transfer_size].copy()
    for scaling in scalings:
        ior_transfer_size_scaling = ior_transfer_size[ior_transfer_size['scaling'] == scaling].copy()
        plt.figure()
        sns.lineplot(x='num_proc', y='time', hue='io_subsystem', data=ior_transfer_size_scaling, style="io_subsystem", markers=True, dashes=False)
        plt.title('iort' + transfer_size + ' ' + scaling + ' runtime over all default io subsystems')
        plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
        plt.savefig('/Users/henryxu/Desktop/paper/papers-2021-prismio/jupyter-notebook/plots/iort' + transfer_size + '/' + 'iort' + transfer_size + '-' + scaling + '-runtime-all-default-io-subsystems.pdf', bbox_inches='tight')
```

### plot ior runtime on each machine

```python
machines = ['lassen', 'summit', 'cori']
transfer_sizes = ['4k', '64k', '1m']
scalings = ['strong-scaling', 'weak-scaling']
for machine in machines:
    ior_machine = ior[ior['io_subsystem'].str.contains(machine)].copy()
    for transfer_size in transfer_sizes:
        ior_machine_transfer_size = ior_machine[ior_machine['transfer_size'] == transfer_size].copy()
        for scaling in scalings:
            ior_machine_transfer_size_scaling = ior_machine_transfer_size[ior_machine_transfer_size['scaling'] == scaling].copy()
            plt.figure()
            sns.lineplot(x='num_proc', y='time', hue='io_subsystem', data=ior_machine_transfer_size_scaling, style="io_subsystem", markers=True, dashes=False)
            plt.title('iort' + transfer_size + ' ' + scaling + ' runtime on ' + machine + 'PFS vs BB')
            plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
            plt.savefig('/Users/henryxu/Desktop/paper/papers-2021-prismio/jupyter-notebook/plots/iort' + transfer_size + '/' + 'iort' + transfer_size + '-' + scaling + '-runtime-' + machine + '.pdf', bbox_inches='tight')
```

### plot ior runtime on cori bb with different transfer sizes 

```python
scalings = ['strong-scaling', 'weak-scaling']
ior_cori = ior[ior['io_subsystem'] == 'cori bb'].copy()
for scaling in scalings:
    ior_cori_scaling = ior_cori[ior_cori['scaling'] == scaling].copy()
    plt.figure()
    sns.lineplot(x='num_proc', y='time', hue='transfer_size', data=ior_cori_scaling, style="transfer_size", markers=True, dashes=False)
    plt.title('ior ' + scaling + ' runtime on Cori BB with different transfer sizes')
    plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
    plt.savefig('/Users/henryxu/Desktop/paper/papers-2021-prismio/jupyter-notebook/plots/ior/ior-cori-bb-' + scaling + '-transfer-size.pdf', bbox_inches='tight')
```
