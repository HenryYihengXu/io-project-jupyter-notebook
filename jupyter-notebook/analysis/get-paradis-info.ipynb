```python
plrw = [plrw64, plrw128, plrw256, plrw512, plrw1024]
plbw = [plbw64, plbw128, plbw256, plbw512, plbw1024]
psrw = [psrw64, psrw128, psrw256, psrw512, psrw1024]
psbw = [psbw64, psbw128, psbw256, psbw512, psbw1024]
pcrw = [pcrw64, pcrw128, pcrw256, pcrw512, pcrw1024]
pcbw = [pcbw64, pcbw128, pcbw256, pcbw512, pcbw1024]

dic = {
    'app': [],
    'io_subsystem': [],
    'scaling': [],
    'num_proc': [],
    'avg_transfer_size_per_read/write': [],
    'min_transfer_size_per_read/write': [],
    'max_transfer_size_per_read/write': [],
    'avg_io_bandwidth_per_read/write': [],
    'min_io_bandwidth_per_read/write': [],
    'max_io_bandwidth_per_read/write': [],
    'avg_io_bandwidth_per_rank': [],
    'min_io_bandwidth_per_rank': [],
    'max_io_bandwidth_per_rank': [],
    'avg_io_time_over_rank': [],
    'min_io_time_over_rank': [],
    'max_io_time_over_rank': [],
    'avg_transfer_time_over_rank': [],
    'min_transfer_time_over_rank': [],
    'max_transfer_time_over_rank': [],
    'avg_io_intensity_over_rank': [],
    'min_io_intensity_over_rank': [],
    'max_io_intensity_over_rank': [],
}

for io_subsystem, ioframe_io_subsystem in zip(
    ['lassen gpfs', 'lassen node-local bb', 'summit gpfs', 'summit node-local bb', 'cori lustre'], 
    [plrw, plbw, psrw, psbw, pcrw, pcbw]):
    for num_proc, recorder_log in zip([64, 128, 256, 512, 1024], ioframe_io_subsystem):
        ioframe = IOFrame(pd.read_pickle(recorder_log + ".pkl"),
                pd.read_pickle(recorder_log + ".meta.pkl")
            )
        print("finish loading")
           
        dic['app'].append('paradis')
        dic['io_subsystem'].append(io_subsystem)
        dic['scaling'].append('weak-scaling')
        dic['num_proc'].append(num_proc)

        avg_transfer_size = ioframe.dataframe['io_volume'].mean()
        min_transfer_size = ioframe.dataframe['io_volume'].min()
        max_transfer_size = ioframe.dataframe['io_volume'].max()
        dic['avg_transfer_size_per_read/write'].append(avg_transfer_size)
        dic['min_transfer_size_per_read/write'].append(min_transfer_size)
        dic['max_transfer_size_per_read/write'].append(max_transfer_size)

        dataframe = ioframe.dataframe[(ioframe.dataframe['io_volume'] > 1) & (ioframe.dataframe['time'] > 0)].copy()
        dataframe['io_bandwidth'] = dataframe['io_volume'] / dataframe['time']
        avg_io_bandwidth = dataframe['io_bandwidth'].mean()
        min_io_bandwidth = dataframe['io_bandwidth'].min()
        max_io_bandwidth = dataframe['io_bandwidth'].max()
        dic['avg_io_bandwidth_per_read/write'].append(avg_io_bandwidth)
        dic['min_io_bandwidth_per_read/write'].append(min_io_bandwidth)
        dic['max_io_bandwidth_per_read/write'].append(max_io_bandwidth)

        io_time = ioframe.io_time(io_type='write,read,meta', agg_function=np.sum, rank_major=True) # agg_column 
        avg_io_time = io_time["time_'write,read,meta'"].mean()
        min_io_time = io_time["time_'write,read,meta'"].min()
        max_io_time = io_time["time_'write,read,meta'"].max()
        dic['avg_io_time_over_rank'].append(avg_io_time)
        dic['min_io_time_over_rank'].append(min_io_time)
        dic['max_io_time_over_rank'].append(max_io_time)

        io_bandwidth_per_rank = ioframe.groupby_aggregate(['rank'], agg_dict={'io_volume': 'sum'})
        io_bandwidth_per_rank = pd.concat([io_bandwidth_per_rank, io_time["time_'write,read,meta'"]], axis=1)
        io_bandwidth_per_rank['io_bandwidth'] = io_bandwidth_per_rank['io_volume'] / io_bandwidth_per_rank["time_'write,read,meta'"]
        avg_io_bandwidth_per_rank = io_bandwidth_per_rank['io_bandwidth'].mean()
        min_io_bandwidth_per_rank = io_bandwidth_per_rank['io_bandwidth'].min()
        max_io_bandwidth_per_rank = io_bandwidth_per_rank['io_bandwidth'].max()
        dic['avg_io_bandwidth_per_rank'].append(avg_io_bandwidth_per_rank)
        dic['min_io_bandwidth_per_rank'].append(min_io_bandwidth_per_rank)
        dic['max_io_bandwidth_per_rank'].append(max_io_bandwidth_per_rank)

        transfer_time = ioframe.io_time(io_type='write,read', agg_function=np.sum, rank_major=True) # agg_column 
        avg_transfer_time = transfer_time["time_'write,read'"].mean()
        min_transfer_time = transfer_time["time_'write,read'"].min()
        max_transfer_time = transfer_time["time_'write,read'"].max()
        dic['avg_transfer_time_over_rank'].append(avg_transfer_time)
        dic['min_transfer_time_over_rank'].append(min_transfer_time)
        dic['max_transfer_time_over_rank'].append(max_transfer_time)

        avg_io_intensity = ioframe.time_distribution(ratio=True)['io_time'].mean()
        min_io_intensity = ioframe.time_distribution(ratio=True)['io_time'].min()
        max_io_intensity = ioframe.time_distribution(ratio=True)['io_time'].max()
        dic['avg_io_intensity_over_rank'].append(avg_io_intensity)
        dic['min_io_intensity_over_rank'].append(min_io_intensity)
        dic['max_io_intensity_over_rank'].append(max_io_intensity)
        
        print("finish processing")
            

df = pd.DataFrame.from_dict(dic)
df.to_csv('/Users/henryxu/Desktop/io-project/data/analysis/paradis-io-info.csv', index=False)
            
```
